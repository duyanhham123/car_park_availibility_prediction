# K-nn algorithm to cluster all carparks in the carpark database

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from sklearn.cluster import KMeans
```

```python
#From stored carpark_info.json, extract data to carpark_info variable and create dataframe carpark_info_df and make a copy
#carpark_copy
import pandas as pd
import json
with open('carpark_info.json', 'r') as f:
    carpark_info = json.load(f)
carpark_info_df = pd.DataFrame(carpark_info)
carpark_copy = carpark_info_df.copy()
```

```python
import pyproj
import utm
import math
#converter function transforms x and y coords in carpark_info_df (or carpark_copy) from SVY21 (a Transverse Mercator projection)
#into latitude and longtitude, and then to WGS84, a coordinate system based on earth's center of mass usedby GPS.
#The reason for doing this is because the database in which postal codes and its relevant info are stored uses WGS84 format
def converter(x_svy,y_svy):
    xfm = pyproj.Transformer.from_crs('EPSG:3414', 'EPSG:4326')
    lat, long = xfm.transform(y_svy, x_svy)
    x, y ,zone,ut= utm.from_latlon(lat, long)
    return (x,y)

#get_disp takes in 3 variables, 
#1. data frame of carpark (i.e. carpark_info_df / carpark_copy)
#2. data frame of postal codes (i.e. df_postal)
#3. string of a postal code of the user's destination
#get_disp then returns a list of the displacements from that postal code to all carparks
def get_disp(df_carpark,df_address,postal_code):
    list_disp = []
    row = df_address[df_address["Postal Code"] == postal_code]
    x_coord, y_coord = row["x"],row["y"]
    for i in range(len(df_carpark)):
        x = float(df_carpark.iloc[i]['x_new'])
        y = float(df_carpark.iloc[i]['y_new'])
        list_disp.append((i,round(math.dist((x_coord,y_coord),(x,y)),2)))
    return list_disp

#This section runs the converter for the x and y coords in the carparks dataframe and appends 2 new cols 
#x_new and y_new that reflect the x and y coords in WGS84 format, similar to that in the postal code df
#This is done so that euclidean distances can be found in metres
x_new = []
y_new = []
for row in range(len(carpark_copy)):
    x_svy = float(carpark_copy.iloc[row]['x_coord'])
    y_svy = float(carpark_copy.iloc[row]['y_coord'])
    converted = converter(x_svy,y_svy)
    x_new.append(converted[0])
    y_new.append(converted[1])
carpark_copy["x_new"] = x_new
carpark_copy["y_new"] = y_new
col_move_x = carpark_copy.pop("x_new")
carpark_copy.insert(5,"x_new",col_move_x)
col_move_y = carpark_copy.pop("y_new")
carpark_copy.insert(6,"y_new",col_move_y)
```

```python
# i loc = locate & collect indexes of rows & cols
coords = carpark_copy.iloc[:, 5:7].values # comma separates rows from cols. <:-1> = index of last col, we dont need it
```

```python
# print(coords) #of array type
# print(coords.shape)
#find max value of x coord
x_max = carpark_copy['x_new'].max()
y_max = carpark_copy['y_new'].max()
x_min = carpark_copy['x_new'].min()
y_min = carpark_copy['y_new'].min()
txt = "Geo coordinates: max X = {}, min x = {}, max y = {}, min y = {}".format(x_max,x_min,y_max,y_min)
print(txt)
```

## We want to tune the cluster number such that the number chosen is able to cover not only all points on the geo coordinate map (i.e covers all carkparks), but has a centroid-to-point distance not exceeding 500m. Rationale: We do not wish to walk more than 500m from our location to the carpark our model classifies our location under.

## Function to find Centroid coordinates

```python
def centroid(*points):
    x_coords = [p[0] for p in points]
    y_coords = [p[1] for p in points]
    _len = len(points)
    centroid_x = sum(x_coords)/_len
    centroid_y = sum(y_coords)/_len
    return [centroid_x, centroid_y]
```

## Cluster number we are testing: 220

```python
cluster_number = 0
avg_dist_list = []
while cluster_number < clusters:
    cluster_df = pd.DataFrame({'xcoord': list(coords[y_kmeans == cluster_number,0]), 
                               'ycoord': list(coords[y_kmeans == cluster_number,1]), 
                               'cluster': [cluster_number+1 for i in range(len(coords[y_kmeans == cluster_number,0]))]}, 
                              columns=['xcoord','ycoord','cluster'])
    if cluster_number == 0:
        combinedcluster_df = cluster_df
        temp_df = cluster_df
    else:
        combinedcluster_df = pd.concat([combinedcluster_df, cluster_df], ignore_index=True, axis = 0)
        temp_df = cluster_df
        
    result = list(temp_df.drop(['cluster'],axis = 1).to_records(index = False))
    cen = list(centroid(*result))
    cent_to_point = []
    for i in range(len(result)):
        cent_to_point.append(math.dist(cen, list(result[i])))
    #average distance between each point and the centroid for ONE cluster:
    avg_dst = sum(cent_to_point) / len(cent_to_point)
    avg_dist_list.append(avg_dst)
    cluster_number+=1

#print(avg_dist_list)
print("Furthest average distance a cluster had between its carparks among all 220 clusters: {}".format(max(avg_dist_list)))
print("Average radius for all 220 clusters: {}".format(sum(avg_dist_list)/len(avg_dist_list)))
```

# On to Machine learning

### The following cell prepares data in this way:
#### Load the datas of the hours of april,may,june,july into a data frame each. For each of the hours, if data was not retrieved by the API due to timed out connection whatsoever, replace the data with the previous hour's data. 
#### For every row in every dataframe, if the lots_availability > lots_total (which does NOT make sense due to whatever reason), then assign lots_availability = lots_total, then, calculate the percentage of availability of lots for every carpark for every hour and insert into a column in each of the dataframes, where the formula for calculation is ( lots_avaliablity / lots_total ) * 100%. 
#### As our project will only target Cars, we drop all the rows that are not data for cars. (type 'C')
#### All the data frames for every hour for all the training months are stored in months_data

```python
#Make monthly data in dataframes
import pandas as pd
import json

months_data = []
with open('april_data.json', 'r') as f:
    april_data = json.load(f)
    months_data.append(april_data)
with open('may_data.json', 'r') as f:
    may_data = json.load(f)
    months_data.append(may_data)
with open('june_data.json', 'r') as f:
    june_data = json.load(f)
    months_data.append(june_data)
with open('july_data.json', 'r') as f:
    july_data = json.load(f)
    months_data.append(july_data)
for month_data in months_data:
    for key in month_data.keys():
        month_data[key] = pd.DataFrame(month_data[key])

#After hours are dataframes, check that if an hour dosent have data, take from the previous timestamp
for month_data in months_data:
    for i in range(len(month_data)):
        if i > 0:
            previous = month_data[str(i-1)]
            current = month_data[str(i)]
            if len(current) == 0:
                month_data[str(i)] = previous
                
#This cell of code checks for if lots available are more than the total lots, which don't make sense
#We handle this by making the available lots = total lots
#We then find the percentage of available lots for all timestamps for all individual carpark numbers.
for month_data in months_data:
    for j in range(len(month_data)):
        lots_percentage = []
        for row_index in range(len(month_data[str(j)])):
            row = month_data[str(j)].iloc[row_index]
            if int(row['lots_available']) > int(row['total_lots']):
                month_data[str(j)].iloc[row_index]['lots_available'] = month_data[str(j)].iloc[row_index]['total_lots']
            try:
                lots_percentage.append((int(row['lots_available'])/int(row['total_lots'] )* 100))
            except ZeroDivisionError:
                lots_percentage.append(0)
        month_data[str(j)]['lots_percentage'] = lots_percentage

#For testing, get all the lots for cars only, and reset the index
for month_data in months_data:
    for k in range(len(month_data)):
        month_data[str(k)] = month_data[str(k)][month_data[str(k)]['lot_type'] == 'C'].reset_index().drop(['index'],axis=1)
```

### Training features and labels for each carpark in a dictionary where key:value = carpark_name: [] , to be filled in later cells.
### Features for ml models are hour and day. Thus we create a list to model the cyclical nature of hours and days in hours_cycle_sine, hours_cycle_cosine, days_cycle_sine and days_cycle_cosine. We also prepare a list of singapore's holidays (dates) for the year 2022 for our project.

```python
import math
import numpy as np
from datetime import date, datetime
import holidays

training_features = {carpark_copy['car_park_no'].iloc[i]:[] for i in range(len(carpark_copy))}
training_labels = {carpark_copy['car_park_no'].iloc[i]:[] for i in range(len(carpark_copy))}

hours_cycle = [i for i in range(24)]
hours_cycle_sine = [np.sin(2 * np.pi * i/24.0) for i in hours_cycle]
hours_cycle_cosine = [np.cos(2 * np.pi * j/24.0) for j in hours_cycle]
days_cycle = [k for k in range(7)]
days_cycle_sine = [np.sin(2 * np.pi * i/7.0) for i in days_cycle]
days_cycle_cosine = [np.cos(2 * np.pi * j/7.0) for j in days_cycle]

  
sg_holidays = list(holidays.Singapore(years = 2022).keys())
for ptr in range(len(sg_holidays)):
    input_holidate = str(sg_holidays[ptr].year) + '/' + str(sg_holidays[ptr].month) + '/' + str(sg_holidays[ptr].day)
    format_holidate = '%Y/%m/%d'
    datetime = datetime.strptime(input_holidate, format_holidate)
    sg_holidays[ptr] = datetime
```

### The cell below does the following: Our group is taking the months of April, May, June and July as training data for our machine learning model. However, since our project is per-carpark,per-hour analysis instead of aggregated data of all carparks per hour, for every carpark, create an array of features and labels with the same length. Each sample feature is a 1x5 matrix with the following 5 features: [sine(hour),cosine(hour),sine(day),cosine(day),holiday(1 or 0)]
### Meanwhile, the label is the lots availability for that carpark for that hour, i.e. 1x1 matrix: [lots_availability in percentage]. 
### Each carpark has its own key:value pair in dictionary training_features and training_labels. The entire training_features dict will have a number of keys equal to the total number of carparks, and so will the labels. 

### Therefore to summarize, given that the total carparks are n and total timestamps are m, the training_features dict will have n keys, where the value of each of the n keys are m x 5, and the training_labels dict will also have n keys, where the value of each of the n keys are m x 1.

```python
#Current training data scope: April, May, June, July
#Current test data scope: Aug
from datetime import datetime
max_timestamps_dict = {'4':30*24-1,'5':31*24-1,'6':30*24-1,'7':31*24-1}
year = '2022'
month_iter = iter(months_data)
for month in max_timestamps_dict.keys(): #This assumes the month in the keys go from 4 - 7 in that order   
    month_data = next(month_iter)
    month_data_keys = month_data.keys()
    for timestamp in month_data_keys:
        data = month_data[timestamp]
        temp_list = []
        day = str((int(timestamp)) // 24 + 1)
        hour_index = (int(timestamp)) % 24
        input_date = year + '/' + month + '/' + day
        format_date = '%Y/%m/%d'
        datetime = datetime.strptime(input_date, format_date)
        day_index = int(datetime.weekday())
        if datetime in sg_holidays:
            holiday = 1
        else:
            holiday = 0
        temp_list.append(hours_cycle_sine[hour_index])
        temp_list.append(hours_cycle_cosine[hour_index])
        temp_list.append(days_cycle_sine[day_index])
        temp_list.append(days_cycle_cosine[day_index])
        temp_list.append(holiday)
        for row in range(len(data)):
            carpark_no = data.iloc[row]['carpark_number']
            lots_percentage = data.iloc[row]['lots_percentage']
            if carpark_no not in list(training_features.keys()):
                training_features[carpark_no] = []
                training_labels[carpark_no] = []
            training_features[carpark_no].append(temp_list)
            training_labels[carpark_no].append(lots_percentage)
        print("Done with timestamp " + timestamp + " of month " + month)
```

# Train an ML model using SVC for EVERY carpark with the training data and store inside ml_dict, where key:value = carpark_name, [model obj, standard scaler for X, standard scaler for Y]

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
ml_dict = {}
for carpark_name in list(training_features.keys()): 
    sc_X = StandardScaler()
    sc_y = StandardScaler()
    y_train = training_labels[carpark_name]
    y_train_reshaped = np.array(y_train).reshape(-1,1)
    svm_X = sc_X.fit_transform(training_features[carpark_name])
    svm_y = sc_y.fit_transform(y_train_reshaped)
    # most important SVR parameter is Kernel type. It can be #linear,polynomial or gaussian SVR. We have a non-linear condition #so we can select polynomial or gaussian but here we select RBF(a #gaussian type) kernel.
    regressor = SVR(kernel='rbf',C = 7)
    regressor.fit(svm_X,svm_y)
    ml_dict[carpark_name] = [regressor,sc_X,sc_y]
#y_pred_list_svr = sc_y.inverse_transform(np.array(regressor.predict(sc_X.transform(test_features['C19M']))).reshape(-1,1))
```
